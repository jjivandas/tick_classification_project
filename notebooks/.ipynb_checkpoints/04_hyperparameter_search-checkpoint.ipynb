## 0. Imports & Config

This notebook finds optimal SVM hyperparameters (C, gamma, kernel) for the
few-shot classification pipeline, then re-runs the full Monte Carlo sweep
with the best parameters for a direct comparison against default settings.

**Approach:**
1. Run GridSearchCV at K=35 across 20 Monte Carlo splits to find best hyperparameters
2. Re-run full sweep (all K values, 100 MC runs) with the optimized parameters
3. Generate comparison plots (default vs. optimized)

```python
# === Imports & Config (mirrors notebook 03, Blocks 0-3) ===

import json, random, hashlib, os, time
from pathlib import Path
from collections import defaultdict
from datetime import datetime

import numpy as np
import pandas as pd
from PIL import Image
import torch
import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, balanced_accuracy_score
from sklearn.model_selection import GridSearchCV, StratifiedKFold

from bioclip.predict import BaseClassifier

# ---- Paths ----
CLASS_NAMES = paths['processed_dir'] / 'class_names_cleaned.json'
DATA        = paths['processed_dir'] / 'final_data_cleaned.json'
EMB_CACHE   = paths['processed_dir'] / 'emb_cache'

# ---- Experiment config ----
SHOTS      = [1, 3, 5, 10, 25, 35]
RUNS       = 100
SEARCH_RUNS = 20          # Monte Carlo runs for hyperparameter search (fewer since GridSearchCV is expensive)
SEARCH_K   = 35           # Only search at the highest shot count
BASE_SEED  = 42
MIN_TOTAL  = 50

# ---- Results directory ----
SVM_RESULTS = paths.get('svm_results_dir', paths['results_dir'] / 'svm')
if isinstance(SVM_RESULTS, str):
    SVM_RESULTS = paths['results_dir'].parent / SVM_RESULTS

RUN_ID = datetime.now().strftime("%y%m%d_%H%M%S")
RUN_DIR = SVM_RESULTS / "runs" / f"{RUN_ID}_hpsearch"
RUN_DIR.mkdir(parents=True, exist_ok=True)

PREDICTIONS_CSV = RUN_DIR / "predictions.csv"
PLOTS_DIR = RUN_DIR / "plots"
PLOTS_DIR.mkdir(parents=True, exist_ok=True)
ANALYSIS_DIR = RUN_DIR / "analysis"
ANALYSIS_DIR.mkdir(parents=True, exist_ok=True)

# ---- Hyperparameter grid ----
PARAM_GRID = {
    'svm__C':      [0.01, 0.1, 1, 10, 100],
    'svm__gamma':  ['scale', 'auto', 0.001, 0.01, 0.1],
    'svm__kernel': ['rbf', 'linear']
}

# Note: 'linear' kernel ignores gamma, so effective combos = 5*5*1 + 5*1*1 = 30 unique
# But GridSearchCV handles this gracefully (gamma is simply unused for linear)

# ---- Seed derivation (same as notebook 03) ----
def seed_for(k, run_id):
    tag = f"{BASE_SEED}_{k}_{run_id}"
    return int(hashlib.sha256(tag.encode()).hexdigest()[:8], 16) % (2**31)

# ---- BioCLIP feature extractor ----
device = "cuda" if torch.cuda.is_available() else "cpu"
BC = BaseClassifier(device=device)

print(f"Device: {device}")
print(f"Run ID: {RUN_ID}_hpsearch")
print(f"Output: {RUN_DIR}")
print(f"Hyperparameter grid: {sum(1 for _ in GridSearchCV(Pipeline([('scaler', StandardScaler()), ('svm', SVC())]), PARAM_GRID, cv=2).get_params()['estimator'].get_params()['steps'])} steps")
print(f"Grid: C={PARAM_GRID['svm__C']}, gamma={PARAM_GRID['svm__gamma']}, kernel={PARAM_GRID['svm__kernel']}")
print(f"Search strategy: {SEARCH_RUNS} MC runs x GridSearchCV(5-fold) at K={SEARCH_K}")
```

## 1. Load Data & Build Specimen Index

Identical to notebook 03 Blocks 1-3.

```python
# === Block 1 — Load Classes and Data -> Build Specimen Index ===

with open(CLASS_NAMES, "r") as f:
    SPECIES_LIST = set(json.load(f))

with open(DATA, "r") as f:
    data = json.load(f)

by_species = defaultdict(lambda: defaultdict(dict))
for r in data:
    sp = r["true_label"]
    if sp not in SPECIES_LIST:
        continue
    sid  = r["sample_id"]
    view = str(r["view"]).strip().lower()
    by_species[sp][sid][view] = r["image_path"]

# Keep only specimens with BOTH views
for sp in list(by_species.keys()):
    for sid in list(by_species[sp].keys()):
        views = by_species[sp][sid]
        if not ("dorsal" in views and "ventral" in views):
            del by_species[sp][sid]
    if not by_species[sp]:
        del by_species[sp]

# Enforce minimum specimen count
for sp in list(by_species.keys()):
    if len(by_species[sp]) < MIN_TOTAL:
        del by_species[sp]

included_species = sorted(by_species.keys())
SPECIES_N = {sp: len(by_species[sp]) for sp in included_species}
CLASS_ORDER = sorted(included_species, key=lambda sp: SPECIES_N[sp], reverse=True)

n_specimens = sum(len(smap) for smap in by_species.values())
print(f"Included species: {len(included_species)}")
print(f"Total usable specimens: {n_specimens}")
for sp in CLASS_ORDER:
    print(f"  {sp}: {SPECIES_N[sp]} specimens")
```

```python
# === Blocks 2-4 — Split, Embedding, and Train/Predict helpers ===

# --- Embedding helpers (Block 3) ---
def _cache_fp(img_path: str) -> Path:
    h = hashlib.sha256(img_path.encode("utf-8")).hexdigest()[:24]
    return EMB_CACHE / f"{h}.npy"

def embed_image(img_path: str) -> np.ndarray:
    fp = _cache_fp(img_path)
    if fp.exists():
        return np.load(fp)
    pil = Image.open(img_path).convert("RGB")
    vec = BC.create_image_features([pil], normalize=True).cpu().numpy()[0]
    np.save(fp, vec)
    return vec

def specimen_vec(rec: dict) -> np.ndarray:
    z_d = embed_image(rec["dorsal"])
    z_v = embed_image(rec["ventral"])
    return 0.5 * (z_d + z_v)

# --- Split helper (Block 2) ---
def split_once(by_species, species_list, k, seed):
    rng = random.Random(seed)
    train_pairs, test_pairs = [], []
    for sp in species_list:
        sids = list(by_species[sp].keys())
        rng.shuffle(sids)
        tr, te = sids[:k], sids[k:]
        assert len(tr) == k, f"{sp}: needs {k} train, found {len(tr)}"
        assert len(te) >= 1, f"{sp}: needs at least 1 test specimen"
        train_pairs.extend([(sp, sid) for sid in tr])
        test_pairs.extend([(sp, sid) for sid in te])
    return train_pairs, test_pairs

def split_for_run(by_species, species_list, k, run_id):
    return split_once(by_species, species_list, k=k, seed=seed_for(k, run_id))

# --- Build X/y (Block 4) ---
def build_xy(by_species, pairs):
    X, y, ids = [], [], []
    for sp, sid in pairs:
        rec = by_species[sp][sid]
        X.append(specimen_vec(rec))
        y.append(sp)
        ids.append(sid)
    return np.stack(X), np.array(y), ids

print("Helpers loaded.")
```

## 2. Hyperparameter Search (GridSearchCV at K=35)

For each of 20 Monte Carlo splits at K=35:
1. Sample training set (35 specimens per species)
2. Run 5-fold stratified GridSearchCV on the training set
3. Record the best parameters and test-set macro accuracy

This finds which (C, gamma, kernel) combination works best across different random splits.

```python
# === Hyperparameter Search ===

from tqdm.auto import tqdm

search_results = []

print(f"Running hyperparameter search: {SEARCH_RUNS} MC runs at K={SEARCH_K}")
print(f"Grid: {len(PARAM_GRID['svm__C'])} C x {len(PARAM_GRID['svm__gamma'])} gamma x {len(PARAM_GRID['svm__kernel'])} kernel = 50 combinations")
print(f"Each combination evaluated with 5-fold stratified CV")
print()

for run_id in tqdm(range(SEARCH_RUNS), desc="HP Search"):
    # 1) Split
    train_pairs, test_pairs = split_for_run(
        by_species, included_species, k=SEARCH_K, run_id=run_id
    )
    
    # 2) Build X/y
    Xtr, ytr, _ = build_xy(by_species, train_pairs)
    Xte, yte, _ = build_xy(by_species, test_pairs)
    
    # 3) GridSearchCV on training set
    pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("svm", SVC(class_weight="balanced", probability=True,
                     random_state=seed_for(SEARCH_K, run_id)))
    ])
    
    cv = StratifiedKFold(n_splits=5, shuffle=True,
                         random_state=seed_for(SEARCH_K, run_id))
    
    grid = GridSearchCV(
        pipe, PARAM_GRID,
        cv=cv,
        scoring='balanced_accuracy',
        n_jobs=-1,
        refit=True
    )
    
    grid.fit(Xtr, ytr)
    
    # 4) Evaluate best model on held-out test set
    yhat = grid.predict(Xte)
    test_macro = balanced_accuracy_score(yte, yhat)
    test_overall = accuracy_score(yte, yhat)
    
    # 5) Also evaluate default params on same split for comparison
    default_pipe = Pipeline([
        ("scaler", StandardScaler()),
        ("svm", SVC(kernel="rbf", class_weight="balanced", probability=True,
                     random_state=seed_for(SEARCH_K, run_id)))
    ])
    default_pipe.fit(Xtr, ytr)
    yhat_default = default_pipe.predict(Xte)
    default_macro = balanced_accuracy_score(yte, yhat_default)
    default_overall = accuracy_score(yte, yhat_default)
    
    search_results.append({
        'run_id': run_id,
        'best_C': grid.best_params_['svm__C'],
        'best_gamma': grid.best_params_['svm__gamma'],
        'best_kernel': grid.best_params_['svm__kernel'],
        'cv_score': grid.best_score_,
        'test_macro_optimized': test_macro,
        'test_overall_optimized': test_overall,
        'test_macro_default': default_macro,
        'test_overall_default': default_overall,
        'improvement': test_macro - default_macro
    })

search_df = pd.DataFrame(search_results)
search_df.to_csv(ANALYSIS_DIR / 'hyperparam_search.csv', index=False)

print("\n" + "=" * 60)
print("HYPERPARAMETER SEARCH RESULTS")
print("=" * 60)
print(f"\nPer-run results (K={SEARCH_K}, {SEARCH_RUNS} MC splits):")
print(search_df[['run_id', 'best_C', 'best_gamma', 'best_kernel',
                  'test_macro_optimized', 'test_macro_default', 'improvement']].to_string(index=False))

print(f"\nMean optimized macro acc: {search_df['test_macro_optimized'].mean():.4f} +/- {search_df['test_macro_optimized'].std():.4f}")
print(f"Mean default macro acc:   {search_df['test_macro_default'].mean():.4f} +/- {search_df['test_macro_default'].std():.4f}")
print(f"Mean improvement:         {search_df['improvement'].mean():.4f}")
```

## 3. Select Best Hyperparameters

Pick the most frequently selected (C, gamma, kernel) combination across the 20 search runs.

```python
# === Select most common best parameters ===

from collections import Counter

param_combos = list(zip(
    search_df['best_C'],
    search_df['best_gamma'].astype(str),
    search_df['best_kernel']
))

combo_counts = Counter(param_combos)
print("Parameter combination frequency:")
for combo, count in combo_counts.most_common():
    print(f"  C={combo[0]}, gamma={combo[1]}, kernel={combo[2]} -> {count}/{SEARCH_RUNS} runs")

best_combo = combo_counts.most_common(1)[0][0]
BEST_C = best_combo[0]
BEST_GAMMA = best_combo[1]
BEST_KERNEL = best_combo[2]

# Convert gamma back from string if needed
if BEST_GAMMA not in ('scale', 'auto'):
    BEST_GAMMA = float(BEST_GAMMA)

print(f"\nSelected parameters (most frequent):")
print(f"  C      = {BEST_C}")
print(f"  gamma  = {BEST_GAMMA}")
print(f"  kernel = {BEST_KERNEL}")
print(f"  (chosen in {combo_counts.most_common(1)[0][1]}/{SEARCH_RUNS} runs)")
```

## 4. Full Monte Carlo Sweep with Optimized Parameters

Re-run the complete sweep (all K values, 100 MC runs) using the best hyperparameters.
This uses the same seeds as notebook 03 for a direct comparison.

```python
# === Full Monte Carlo Sweep with Optimized Params ===

from tqdm.auto import tqdm

print(f"Full sweep with optimized params: C={BEST_C}, gamma={BEST_GAMMA}, kernel={BEST_KERNEL}")
print(f"Shots: {SHOTS} | Runs: {RUNS}")
print()

all_pred_rows = []
failed_runs = []

for shots in SHOTS:
    pbar = tqdm(range(RUNS), desc=f"{shots}-shot", leave=True)
    for run_id in pbar:
        try:
            seed = seed_for(shots, run_id)
            train_pairs, test_pairs = split_for_run(
                by_species, included_species, k=shots, run_id=run_id
            )
            
            Xtr, ytr, _ = build_xy(by_species, train_pairs)
            Xte, yte, te_ids = build_xy(by_species, test_pairs)
            
            # Optimized SVM
            clf = Pipeline([
                ("scaler", StandardScaler()),
                ("svm", SVC(C=BEST_C, gamma=BEST_GAMMA, kernel=BEST_KERNEL,
                            class_weight="balanced", probability=True,
                            random_state=seed))
            ])
            
            clf.fit(Xtr, ytr)
            yhat = clf.predict(Xte)
            probs = clf.predict_proba(Xte)
            conf = probs.max(axis=1)
            
            overall_acc = accuracy_score(yte, yhat)
            macro_acc = balanced_accuracy_score(yte, yhat)
            
            for i, (sp_true, sid) in enumerate(zip(yte, te_ids)):
                all_pred_rows.append({
                    'shots': shots,
                    'run_id': run_id,
                    'sample_id': sid,
                    'species_true': sp_true,
                    'species_pred': yhat[i],
                    'correct': int(sp_true == yhat[i]),
                    'confidence': float(conf[i]),
                    'macro_acc': macro_acc,
                    'overall_acc': overall_acc
                })
            
            pbar.set_postfix(macro=f"{macro_acc:.3f}", overall=f"{overall_acc:.3f}")
        
        except Exception as e:
            failed_runs.append((shots, run_id, str(e)))
            print(f"[ERROR] shots={shots} run_id={run_id} | {e}")

# Save predictions
df_all = pd.DataFrame(all_pred_rows)
df_all.to_csv(PREDICTIONS_CSV, index=False)

print(f"\nSweep complete. {len(df_all)} rows saved to {PREDICTIONS_CSV}")
if failed_runs:
    print(f"Failures: {len(failed_runs)}")
```

## 5. Aggregate Metrics & Comparison

```python
# === Aggregate metrics per shot count ===

# Optimized results (from this notebook)
opt_metrics = []
for K in SHOTS:
    subset = df_all[df_all['shots'] == K]
    run_macros = subset.groupby('run_id')['macro_acc'].first()
    run_overalls = subset.groupby('run_id')['overall_acc'].first()
    opt_metrics.append({
        'shots': K,
        'mean_macro': run_macros.mean(),
        'std_macro': run_macros.std(),
        'ci95_macro': 1.96 * run_macros.std() / np.sqrt(len(run_macros)),
        'mean_overall': run_overalls.mean(),
        'std_overall': run_overalls.std(),
        'ci95_overall': 1.96 * run_overalls.std() / np.sqrt(len(run_overalls)),
        'n_runs': len(run_macros)
    })

opt_summary = pd.DataFrame(opt_metrics)
opt_summary.to_csv(ANALYSIS_DIR / 'shot_summary.csv', index=False)

# Per-species accuracy
species_rows = []
for K in SHOTS:
    subset = df_all[df_all['shots'] == K]
    for sp in CLASS_ORDER:
        sp_data = subset[subset['species_true'] == sp]
        per_run = sp_data.groupby('run_id')['correct'].mean()
        species_rows.append({
            'shots': K,
            'species': sp,
            'acc_mean': per_run.mean(),
            'acc_std': per_run.std(),
            'acc_ci95': 1.96 * per_run.std() / np.sqrt(len(per_run)),
            'n_runs': len(per_run)
        })

species_df = pd.DataFrame(species_rows)
species_df.to_csv(ANALYSIS_DIR / 'per_species_accuracy.csv', index=False)

print("Optimized results (shot summary):")
print(opt_summary[['shots', 'mean_macro', 'std_macro', 'mean_overall', 'n_runs']].to_string(index=False))
```

```python
# === Load default results for comparison ===
# Point this to the most recent default-params run

DEFAULT_RUN = SVM_RESULTS / "runs" / "260324_164203" / "analysis" / "shot_summary.csv"

if DEFAULT_RUN.exists():
    default_summary = pd.read_csv(DEFAULT_RUN)
    print("Default params results loaded.")
else:
    print(f"WARNING: Default run not found at {DEFAULT_RUN}")
    print("Comparison plots will be skipped.")
    default_summary = None

# Build comparison table
if default_summary is not None:
    comp = opt_summary[['shots', 'mean_macro', 'mean_overall']].copy()
    comp.columns = ['shots', 'optimized_macro', 'optimized_overall']
    
    merged = comp.merge(
        default_summary[['shots', 'mean_macro', 'mean_overall']].rename(
            columns={'mean_macro': 'default_macro', 'mean_overall': 'default_overall'}
        ),
        on='shots', how='left'
    )
    merged['macro_diff'] = merged['optimized_macro'] - merged['default_macro']
    merged['overall_diff'] = merged['optimized_overall'] - merged['default_overall']
    
    merged.to_csv(ANALYSIS_DIR / 'default_vs_optimized.csv', index=False)
    
    print("\nDefault vs. Optimized Comparison:")
    print("=" * 80)
    for _, row in merged.iterrows():
        K = int(row['shots'])
        dm = row.get('default_macro', float('nan'))
        om = row['optimized_macro']
        diff = row.get('macro_diff', float('nan'))
        sign = '+' if diff >= 0 else ''
        print(f"  K={K:2d}  default={dm*100:5.1f}%  optimized={om*100:5.1f}%  diff={sign}{diff*100:.1f}%")
    
    print(f"\nOptimized params: C={BEST_C}, gamma={BEST_GAMMA}, kernel={BEST_KERNEL}")
    print(f"Default params:   C=1.0, gamma=scale, kernel=rbf")
```

## 6. Comparison Plots

```python
# === Macro Accuracy Curve: Default vs. Optimized ===

if default_summary is not None:
    fig, ax = plt.subplots(figsize=(10, 6))
    
    # Default params curve
    d = default_summary
    ax.plot(d['shots'], d['mean_macro'] * 100, 'o--', color='gray', linewidth=1.5,
            markersize=6, label='Default (C=1, gamma=scale, rbf)', alpha=0.7)
    ax.fill_between(d['shots'],
                    (d['mean_macro'] - d['ci95_macro']) * 100,
                    (d['mean_macro'] + d['ci95_macro']) * 100,
                    alpha=0.1, color='gray')
    
    # Optimized params curve
    o = opt_summary
    ax.plot(o['shots'], o['mean_macro'] * 100, 'o-', color='#1f77b4', linewidth=2.5,
            markersize=8, label=f'Optimized (C={BEST_C}, gamma={BEST_GAMMA}, {BEST_KERNEL})')
    ax.fill_between(o['shots'],
                    (o['mean_macro'] - o['ci95_macro']) * 100,
                    (o['mean_macro'] + o['ci95_macro']) * 100,
                    alpha=0.15, color='#1f77b4')
    
    # Annotate optimized points
    for _, row in o.iterrows():
        ax.annotate(f"{row['mean_macro']*100:.1f}%",
                    (row['shots'], row['mean_macro']*100),
                    textcoords='offset points', xytext=(0, 12),
                    fontsize=9, fontweight='bold', ha='center', color='#1f77b4')
    
    ax.set_xlabel('Shots per species', fontsize=12)
    ax.set_ylabel('Macro accuracy', fontsize=12)
    ax.set_title('Macro Accuracy vs. Shots: Default vs. Optimized SVM Hyperparameters\n'
                 f'(mean \u00b1 95% CI over {RUNS} runs)', fontsize=13)
    ax.legend(loc='lower right', fontsize=10)
    ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0f}%'))
    ax.grid(alpha=0.3)
    ax.set_ylim(50, 100)
    
    plt.tight_layout()
    plt.savefig(PLOTS_DIR / 'hyperparam_comparison.png', dpi=150, bbox_inches='tight')
    plt.show()
    
    print(f"Saved: {PLOTS_DIR / 'hyperparam_comparison.png'}")
else:
    print("Skipped comparison plot (no default run found).")
```

```python
# === Standalone Macro Accuracy Curve (optimized params only) ===

fig, ax = plt.subplots(figsize=(10, 6))

o = opt_summary
ax.plot(o['shots'], o['mean_macro'] * 100, 'o-', color='#1f77b4', linewidth=2.5, markersize=8,
        label=f'BioCLIP 2 (SVM, C={BEST_C}, {BEST_KERNEL})')
ax.fill_between(o['shots'],
                (o['mean_macro'] - o['ci95_macro']) * 100,
                (o['mean_macro'] + o['ci95_macro']) * 100,
                alpha=0.15, color='#1f77b4')

for _, row in o.iterrows():
    ax.annotate(f"{row['mean_macro']*100:.1f}%",
                (row['shots'], row['mean_macro']*100),
                textcoords='offset points', xytext=(0, 12),
                fontsize=10, fontweight='bold', ha='center')

ax.set_xlabel('Shots per species', fontsize=12)
ax.set_ylabel('Macro accuracy', fontsize=12)
ax.set_title(f'Macro Accuracy vs. Shots (mean \u00b1 95% CI over {RUNS} runs)\n'
             f'Optimized: C={BEST_C}, gamma={BEST_GAMMA}, kernel={BEST_KERNEL}', fontsize=13)
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0f}%'))
ax.grid(alpha=0.3)
ax.set_ylim(50, 100)

plt.tight_layout()
plt.savefig(PLOTS_DIR / 'macro_accuracy_curve.png', dpi=150, bbox_inches='tight')
plt.show()

print(f"Saved: {PLOTS_DIR / 'macro_accuracy_curve.png'}")
```

## 7. Summary

```python
# === Final Summary ===

print("=" * 60)
print("HYPERPARAMETER OPTIMIZATION SUMMARY")
print("=" * 60)
print(f"\nDefault parameters:  C=1.0, gamma=scale, kernel=rbf")
print(f"Optimized parameters: C={BEST_C}, gamma={BEST_GAMMA}, kernel={BEST_KERNEL}")
print(f"\nSearch method: GridSearchCV (5-fold stratified CV)")
print(f"Search runs: {SEARCH_RUNS} Monte Carlo splits at K={SEARCH_K}")
print(f"Full evaluation: {RUNS} Monte Carlo runs across K={SHOTS}")
print()

print("Optimized Macro Accuracy:")
for _, row in opt_summary.iterrows():
    print(f"  K={int(row['shots']):2d}: {row['mean_macro']*100:5.1f}% +/- {row['ci95_macro']*100:.1f}%")

if default_summary is not None:
    print("\nComparison (optimized - default):")
    comp = pd.read_csv(ANALYSIS_DIR / 'default_vs_optimized.csv')
    for _, row in comp.iterrows():
        diff = row['macro_diff'] * 100
        sign = '+' if diff >= 0 else ''
        print(f"  K={int(row['shots']):2d}: {sign}{diff:.2f} percentage points")

print(f"\nAll results saved to: {RUN_DIR}")
print(f"  - {ANALYSIS_DIR / 'hyperparam_search.csv'}")
print(f"  - {ANALYSIS_DIR / 'shot_summary.csv'}")
print(f"  - {ANALYSIS_DIR / 'per_species_accuracy.csv'}")
print(f"  - {ANALYSIS_DIR / 'default_vs_optimized.csv'}")
print(f"  - {PLOTS_DIR / 'hyperparam_comparison.png'}")
print(f"  - {PLOTS_DIR / 'macro_accuracy_curve.png'}")
```
